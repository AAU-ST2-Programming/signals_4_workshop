# **Øvelse 1 (2t): "Komplet signalanalyse-pipeline fra Patient #10"**

## Scenarie
Du modtager EKG (4 leads) og PPG fra to tidspunkter hos Patient #10.  
Du skal bygge en analyse-pipeline fra rådata til klinisk fortolkning.  

**Datafiler (signals_4_workshop/files):**  
- `10_18-12-33_ecg.csv` (lead_I, lead_II, v3, v4)  
- `10_18-12-33_ppg.csv`  
- Identiske filer fra tidspunkt 18:13-23  

## Du ved
- CSV-indlæsning med pandas  
- Mean, std, min, max for tidsserier  
- Plots i tidsdomæne og histogrammer  
- Low-pass filtrering (glidende gennemsnit eller Butterworth)  
- Features som måltal der opsummerer signalet  

## Opgave
Byg en reproducerbar pipeline der indlæser EKG og PPG.  
Visualiser dem, filtrer støj og uddrag relevante features.  
Saml alle features i en tabel med metadata.  
Gem resultater til CSV og plots.  

## Krav
- Pandas til indlæsning og tabeller  
- Matplotlib/seaborn til plots  
- Funktioner: `load_signal()`, `filter_signal()`, `extract_features()`  
- Docstrings for alle funktioner  
- Reproducerbar (samme output hver gang)  
- Kort klinisk fortolkning af resultaterne  
- Forslag til filnavne: `exercise1_pipeline.py`, `exercise1_features.csv`, `exercise1_report.md`  

# **Øvelse 2 (2t): "OOP-refaktorering + regression-analyse dag-til-dag"**

## Scenarie
Pipeline-koden fra Øvelse 1 skal gøres genbrugelig.  
Patienten følges over tid.  
Du skal refaktorere til OOP for EKG og PPG.  
Du skal sammenligne Day 1 vs Day 2 med regression (y = m*x + b).  

## Du ved
- CSV-parsing (fra Øvelse 1)  
- Klasser, metoder, indkapsling (forelæsning 2)  
- Lineær regression: y = m*x + b, R² værdi  
- Scatter plots og fit  
- Klinisk fortolkning  

## Opgave
Refaktorér Øvelse 1-koden til OOP-klasser (`Signal`-baseklasse + ECGSignal og PPGSignal subclasses).  
Brug klasserne til at indlæse, filtrere og uddrage features for begge dage.  
Brug regressionens parametre og kvalitet til at vurdere om en feature er stabil eller ændrer sig dag til dag.  

**Hvad klasserne skal gøre:**  
- Indlæse CSV, filtrere data, uddrage features, returnere data  
- Private data  
- Exceptions ved ugyldige data  

## Krav
- Klasser i separat `.py`-modul (ECGSignal og PPGSignal)  
- Docstrings for alle metoder  
- Regression metode: fit_regression() eller lignende  
- Regressionsresultater-tabel: `m`, `b`, `R²` for hver feature  
- Forslag til filnavne: `exercise2_signals.py`, `exercise2_analysis.py`, `exercise2_regression.csv`
